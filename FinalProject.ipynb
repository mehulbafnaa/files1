```python

import requests
import pandas as pd
from urllib.parse import quote
import pickle
import os

# Function to save data to a file
def save_data_to_cache(file_name, data):
    with open(file_name, 'wb') as f:
        pickle.dump(data, f)

# Function to load data from a file
def load_data_from_cache(file_name):
    if os.path.exists(file_name):
        with open(file_name, 'rb') as f:
            return pickle.load(f)
    return None

# Set up the API request parameters
api_key = "E65L6Q4DW4FNKRDD6AMLN9BAU"  # actual API key
location = quote("New York City, NY")  # URL encode the location
start_date = "2022-01-01"
end_date = "2022-12-31"
unit_group = "us"  # Use  'us' for Fahrenheit

# Check if the data is already cached
cache_file = f"weather_data_{location}_{start_date}_to_{end_date}.pkl"
weather_data = load_data_from_cache(cache_file)

if weather_data is None:
    # Data is not cached, fetch from API
    url = f"https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{location}/{start_date}/{end_date}?unitGroup={unit_group}&include=days&key={api_key}&contentType=json"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        weather_data = []
        for day in data.get('days', []):
            date = day.get('datetime')
            max_temp = day.get('tempmax')
            min_temp = day.get('tempmin')
            avg_temp = day.get('temp')
            conditions = day.get('conditions')
            precipitation = day.get('precip', 0)  # Default to 0 if not present
            wind_speed = day.get('windspeed')
            snow = day.get('snow', 0)  # Default to 0 if not present

            weather_data.append([date, max_temp, min_temp, avg_temp, conditions, precipitation, wind_speed, snow])

        # Cache the fetched data
        save_data_to_cache(cache_file, weather_data)
    else:
        print("Failed to retrieve data:", response.status_code)
        weather_data = []

# Create a Pandas DataFrame from the weather data
df_weather = pd.DataFrame(weather_data, columns=['Date', 'Max Temperature (°F)', 'Min Temperature (°F)',
                                         'Average Temperature (°F)', 'Conditions', 'Precipitation (inches)',
                                         'Wind Speed (mph)', 'Snow (inches)'])

# Display the DataFrame
display(df_weather.head())



```

```python
import pandas as pd

# Assuming df_combined is your DataFrame

# Function to calculate missing values by column
def missing_values(data):
    nan_counts = data.isna().sum()
    null_counts = data.apply(lambda x: (x == '(null)').sum() if x.dtype == object else 0)

    # Safely apply string methods only to string columns
    unknown_counts = data.apply(lambda x: (x.astype(str).str.lower() == 'unknown').sum() if pd.api.types.is_string_dtype(x) else 0)

    # Creating a DataFrame to hold the results
    results = pd.DataFrame({
        'NaN Count': nan_counts,
        '(null) Count': null_counts,
        'Unknown Count': unknown_counts
    })
    return results

# Calling the function and storing the result
missing_data = missing_values(df_combined)

# Display the results
print("Missing Values Count per Column:")
print(missing_data)
```

```python
import psycopg2
from psycopg2.extras import execute_values

# Database connection parameters
conn_params = {
    "dbname": "postgres",
    "user": "postgres",
    "password":"postgres",
    "host": "localhost" ,
    # or your database server address
}

# Connect to the database
conn = psycopg2.connect(**conn_params)
cur = conn.cursor()

# Prepare the data for insertion
# Convert DataFrame to a list of tuples
data_tuples = list(df_combined.itertuples(index=False, name=None))

# SQL query for inserting data
insert_query = """
INSERT INTO complaints (
    CMPLNT_NUM, CMPLNT_FR_DT, CMPLNT_FR_TM, CMPLNT_TO_DT, CMPLNT_TO_TM,
    ADDR_PCT_CD, RPT_DT, KY_CD, OFNS_DESC, PD_CD, PD_DESC, CRM_ATPT_CPTD_CD,
    LAW_CAT_CD, BORO_NM, LOC_OF_OCCUR_DESC, PREM_TYP_DESC, JURIS_DESC,
    JURISDICTION_CODE, PARKS_NM, HADEVELOPT, HOUSING_PSA, X_COORD_CD, Y_COORD_CD,
    SUSP_AGE_GROUP, SUSP_RACE, SUSP_SEX, TRANSIT_DISTRICT, Latitude, Longitude,
    Lat_Lon, PATROL_BORO, STATION_NAME, VIC_AGE_GROUP, VIC_RACE, VIC_SEX,
    Max_Temperature_F, Min_Temperature_F, Average_Temperature_F, Conditions,
    Precipitation_inches, Wind_Speed_mph, Snow_inches
) VALUES %s
"""

try:
    # Execute the insertion using execute_values for efficient bulk insert
    execute_values(cur, insert_query, data_tuples)
    conn.commit()  # Commit the transaction
    print("Data inserted successfully.")
except Exception as e:
    print(f"An error occurred: {e}")
finally:
    cur.close()
    conn.close()







```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller, acf, pacf
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.statespace.sarimax import SARIMAX
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# df_combined = pd.read_csv('your_data.csv')

# Step 1: Aggregate crime counts by date
crime_counts = df_combined['CMPLNT_FR_DT'].value_counts().sort_index()

# Step 2: Aggregate weather data by date
daily_weather = df_combined.groupby('CMPLNT_FR_DT')[['Max Temperature (°F)', 'Precipitation (inches)', 'Wind Speed (mph)', 'Snow (inches)']].mean()

# Combine crime data with weather data
daily_data = pd.concat([crime_counts, daily_weather], axis=1)
daily_data.columns = ['Crime Count', 'Avg Temp (°F)', 'Precipitation (inches)', 'Wind Speed (mph)', 'Snow (inches)']

# Remove any NaN values that may affect the analysis
daily_data.dropna(inplace=True)

# Step 3: Test for stationarity
adf_result = adfuller(daily_data['Crime Count'])
print(f'ADF Statistic: {adf_result[0]}')
print(f'p-value: {adf_result[1]}')

# If data is not stationary, perform differencing
if adf_result[1] > 0.05:
    daily_data['Crime Count Differenced'] = daily_data['Crime Count'].diff().dropna()

# Plot ACF and PACF to determine the order of the ARIMA model
plot_acf(daily_data['Crime Count Differenced'].dropna(), lags=40)
plt.show()

plot_pacf(daily_data['Crime Count Differenced'].dropna(), lags=40)
plt.show()

# Determine the p, d, q values from the plots
# For example purposes, let's say they are p=1, d=1, q=1
p = 1
d = 1
q = 1

# Step 4: Fit the ARIMA model
arima_model = ARIMA(daily_data['Crime Count Differenced'], order=(p, d, q))
arima_result = arima_model.fit()

# Print out the summary of the model
print(arima_result.summary())

# Plot the residuals
residuals = arima_result.resid
plt.figure(figsize=(10, 4))
plt.plot(residuals)
plt.title('Residuals from ARIMA Model')
plt.show()

# Conduct any additional diagnostics on the residuals

# If weather data is to be included in the model as exogenous variables, consider SARIMAX
# For demonstration purposes, I'm using the same p, d, q values
sarimax_model = SARIMAX(daily_data['Crime Count'],
                         order=(p, d, q),
                         exog=daily_data[['Avg Temp (°F)', 'Precipitation (inches)', 'Wind Speed (mph)', 'Snow (inches)']])
sarimax_result = sarimax_model.fit()

# Print out the summary of the SARIMAX model
print(sarimax_result.summary())

# Plot the SARIMAX residuals
sarimax_residuals = sarimax_result.resid
plt.figure(figsize=(10, 4))
plt.plot(sarimax_residuals)
plt.title('Residuals from SARIMAX Model')
plt.show()

# # Additional diagnostics for SARIMAX residuals
```
